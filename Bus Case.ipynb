```python
import pandas as pd
import statsmodels.api as sm
```

```python
data = pd.read_csv('busscase.txt', header = None,) 
cols = ['ZON', 'BUSSALD', 'KUNDNR', 'ANTAVT', 'DUR', 'ANTSKAD', 'SKADKOST']
df = pd.DataFrame(data, columns = cols)
for i in range(len(data)):
    temp = data.iloc[i].str.split()[0]
    for j in range(len(temp)):
        try :
            df[cols[j]].iloc[i] = int(temp[j])
        except :
            df[cols[j]].iloc[i] = temp[j]
df
```
